### 01) WAP to find sum of all the elements in List.

```python
l1 = [];
size = int(input("Enter size of list : "));
for i in range(0,size):
  l1.append((int(input("Enter value : "))));
print(f"Sum of list elements : {sum(l1)}");
```

### 02) WAP to find largest element in a List.

```python
l1 = [];
size = int(input("Enter size of list : "));
for i in range(0,size):
  l1.append((int(input("Enter value : "))));
print(f"Largest element is {max(l1)}");
```

### 05) WAP to interchange the elements on two positions entered by a user.

```python
l1 = enterList();
print(f"=======> Enter index between 0 to {len(l1)} <=======");
index1 = int(input("Enter index-1 : "));
index2 = int(input("Enter index-2 : "));

if (index1 >= 0 and index1 <= len(l1)) and (index2 >= 0 and index2 <= len(l1)):
  l1[index1], l1[index2] = l1[index2], l1[index1];
  print(f"Updated List = {l1}");
else:
  print("Enter index between 0 to lenght of list");
```

### 07) Python program to remove multiple elements from a list using list comprehension

```python
l1 = enterList();
print("Enter indexes to remove =>");
l2 = enterList();
if len(l2) < len(l1):
  l3 = [l1[i] for i in range(len(l1)) if i not in l2];
  print(f"Original list = {l1}");
  print(f"Removed elements' indexs = {l2}");
  print(f"Updated list = {l3}");
else:
  print(f"Enter indexs in 0 to {len(l1)-1}");
```

### 08) Create a list from the specified start to end index of another list.

```python
l1 = enterList();
l2 = [];
print(f"=======> Enter index between 0 to {len(l1)} <=======");
index1 = int(input("Starting index : "));
index2 = int(input("Ending index : "));

if (index1 >= 0 and index1 <= len(l1)) and (index2 >= 0 and index2 <= len(l1)):
  l2.extend(l1[index1-1: index2]);
  print(l2);
else:
  print("Enter index between 0 to lenght of list");
```

### 01) WAP to count Even and Odd numbers in a List.

```python
l1 = enterList();
even = 0;
odd = 0;
for i in range(len(l1)):
  if l1[i]%2 == 0:
    even+=1;
  else:
    odd+=1;
print(f"Total even : {even}");
print(f"Total odd : {odd}");
```

### 02) Python program to find N largest and smallest elements from the list

```python
l1 = enterList();

n = int(input("Enter no of elements to get largest and smallest : "));

if n < len(l1):
  descList = sorted(l1, reverse=True);
  maximum = descList[:n];

  ascList = sorted(l1);
  minimum = ascList[:n];
else:
  print(f"Enter N in 0 to {len(l1)}");
print(f"Maximum = {maximum}");
print(f"Minimum = {minimum}");
```

### 03) WAP to print duplicates from a list of integers

```python
l1 = enterList();
l2 = [];
for i in range(len(l1)):
  for j in range(i+1, len(l1)):
    if l1[i] == l1[j]:
      l2.append(l1[i]);
      break;
print(f"Duplicates = {l2}");
```

```python
def enterList():
  size = int(input("Enter size of list : "));
  l1 = [];
  for i in range(0,size):
    l1.append((int(input("Enter value : "))));
  return l1;
```
